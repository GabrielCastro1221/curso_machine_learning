```python
import pandas as pd
from sklearn.svm import SVR
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
```

### _***Maquina de soporte vectorial***_.

+ Se carga el conjunto de datos y se elimina la columna _***Unnamed: 0***_.

```python
data = pd.read_csv("../dataSets/advertising.csv")
data = data.iloc[:, 1:]
data.head()
```

+ Separar la variable dependiente _***X***_ y la variable independiente _***Y***_
+ Se le pasan las variables de entrenamineto tanto de _***'X'***_ como de _***'Y'***_ para crear la maquina de soporte vectorial.

```python
x = data.drop(["Radio", "Sales"], axis=1).values
y = data["Sales"].values

x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size=0.2, random_state=42
)
```

+ Crear una instancia de SVR para crear la maquina de soporte vectorial

```python
svr = SVR(kernel="rbf")
svr.fit(x_train, y_train)
```

+ Se hacen las predicciones.

```python
y_pred = svr.predict(x_test)
print("Reales:", y_test[:4], "Prediccion :", y_pred[:4])
```

+ Se calcula la R2 para saber que tanta presicion tiene la maquina de soporte vectorial.

```python
r2_score(y_test, y_pred)
```

+ Analizando las predicciones de la egresion lineal de este mismo caso se puede observar que la combinacion entre TV y Newspaper sigue siendo mala, por lo que la publicidad en el medio de comunicacion Newspaper se deberia eliminar del sistema de la empresa.

#### A continuacion se haran las predicciones con las varibales de _***TV***_ y _***Radio***_

```python
x = data.drop(["Newspaper", "Sales"], axis=1).values
y = data["Sales"].values

x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size=0.2, random_state=42
)
```

```python
svr = SVR(kernel="rbf")
svr.fit(x_train, y_train)
```

```python
y_pred = svr.predict(x_test)
print("Reales:", y_test[:4], "Prediccion :", y_pred[:4])
```

```python
r2_score(y_test, y_pred)
```

#### Con respecto al modelo de regresion lineal simple no hay mucho cambio en la acertividad de la prediccion
